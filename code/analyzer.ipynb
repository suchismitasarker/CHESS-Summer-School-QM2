```python
import os
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

df = pd.read_csv("percentile2.csv")
df = df.sort_values(by="temp")

output_dir = "percentile2plots"
os.makedirs(output_dir, exist_ok=True)

# Identify all numeric columns that we want to plot on the Y-axis
exclude_cols = [
    "temp",
    "shape",
    "total_pixels",
    "min_val",
    "percentiles",
    "memory_mb",
    "log_min",
    "log_max",
    "nonzero_min",
    "index",
    "path",
    "sample_name",
]
y_columns = [col for col in df.columns if col not in exclude_cols]

print(f"Generating {len(y_columns)} sets of plots...")

# Loop through each quantity and generate the 3 plots side-by-side
for col in y_columns:
    # Use subplots to create a 1-row, 3-column layout
    fig, axes = plt.subplots(1, 3, figsize=(18, 5))

    # --- Graph 1: Original Plot (Dots connected by a line) ---
    axes[0].plot(
        df["temp"],
        df[col],
        marker="o",
        linestyle="-",
        color="b",
        alpha=0.7,
    )
    axes[0].set_xlabel("Temperature (K)", fontsize=12)
    axes[0].set_ylabel(col, fontsize=12)
    axes[0].set_title("Line with Points", fontsize=12)
    axes[0].grid(True, linestyle="--", alpha=0.5)

    # --- Graph 2: Only Points (Scatter Plot) ---
    axes[1].plot(
        df["temp"],
        df[col],
        marker="o",
        linestyle="None",
        color="b",
        alpha=0.7,
    )
    axes[1].set_xlabel("Temperature (K)", fontsize=12)
    axes[1].set_ylabel(col, fontsize=12)
    axes[1].set_title("Only Points", fontsize=12)
    axes[1].grid(True, linestyle="--", alpha=0.5)

    x = df["temp"]
    y = df[col]

    # Calculate slope, intercept, and Pearson correlation coefficient (r)
    slope, intercept = np.polyfit(x, y, 1)
    r_matrix = np.corrcoef(x, y)
    r_coef = r_matrix[0, 1] if r_matrix.shape == (2, 2) else np.nan

    # Plot the data points
    axes[2].plot(
        x,
        y,
        marker="o",
        linestyle="None",
        color="b",
        alpha=0.7,
        label="Data",
    )
    # Plot the line of best fit
    axes[2].plot(
        x,
        slope * x + intercept,
        color="r",
        linestyle="--",
        label=f"Fit ($r$ = {r_coef:.3f}, $m$ = {slope:.3f}, $b$= {intercept:.3f})",
    )
    axes[2].legend(loc="best")
    axes[2].set_xlabel("Temperature (K)", fontsize=12)
    axes[2].set_ylabel(col, fontsize=12)
    axes[2].set_title("Line of Best Fit", fontsize=12)
    axes[2].grid(True, linestyle="--", alpha=0.5)

    # Add a main overarching title for the set of plots
    fig.suptitle(f"Temperature dependence of {col}", fontsize=15, y=0.98)

    # Clean layout and save
    plt.tight_layout(rect=[0, 0, 1, 0.95])
    safe_filename = f"{col.lower().replace(' ', '_')}_vs_temp_combined.png"
    plt.savefig(os.path.join(output_dir, safe_filename), dpi=150)
    plt.close()

print(f"Done! Check the '{output_dir}' folder for all your generated graphs.")
```
